```python
#import differen tlibraries
from scipy.signal import convolve2d
from skimage.data import coffee
from skimage import color
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# image of the data set
plt.imshow(coffee())
plt.show()
```

```python
#converting to greyscale
img = color.rgb2gray(coffee())
plt.imshow(img,cmap=plt.cm.gray)
plt.axis('off')
plt.show()
```

```python
# filters/kernels for edge detection

kernel_edge_detect1 = np.array([[1.,0.,-1.],
                               [0.,0.,0.],
                               [-1.,0.,1.]])

kernel_edge_detect2 = np.array([[0.,1.,0.],
                               [1.,-4.,1.],
                               [0.,1.,0.]])

kernel_edge_detect3 = np.array([[-1.,-1.,-1.],
                               [-1.,8.,-1.],
                               [-1.,-1.,-1.]])

# filters/kernels for sharpen image

kernel_sharp_detect1 = np.array([[0.,-1.,0.],
                               [-1.,5.,-1.],
                               [0.,-1.,0.]])

kernel_sharp_detect2 = np.array([[-1.,-1.,-1.],
                               [-1.,9.,-1.],
                               [-1.,-1.,-1.]])

# filers/kernels for blur image

kernel_blur_detect = np.array([[1.,1.,1.],
                               [1.,1.,1.],
                               [1.,1.,1.]])

```

```python
img_edge1 = convolve2d(img, kernel_edge_detect1, 'valid')
plt.imshow(img_edge1, cmap=plt.cm.gray)
plt.axis('off')
plt.show()
```

```python
img_edge2 = convolve2d(img, kernel_edge_detect2, 'valid')
plt.imshow(img_edge2, cmap=plt.cm.gray)
plt.axis('off')
plt.show()
```

```python
img_edge3 = convolve2d(img, kernel_edge_detect3, 'valid')
plt.imshow(img_edge3, cmap=plt.cm.gray)
plt.axis('off')
plt.show()
```

```python
img_sharp1 = convolve2d(img, kernel_sharp_detect1, 'valid')
plt.imshow(img_sharp1, cmap=plt.cm.gray)
plt.axis('off')
plt.show()
```

```python
img_sharp2 = convolve2d(img, kernel_sharp_detect2, 'valid')
plt.imshow(img_sharp2, cmap=plt.cm.gray)
plt.axis('off')
plt.show()
```

```python
img_blur = convolve2d(img, kernel_blur_detect, 'valid')
plt.imshow(img_blur, cmap=plt.cm.gray)
plt.axis('off')
plt.show()
```
